# **Dimensional Modeling**
Dimensional modeling is a technique designed to optimize data query and analysis. It differs from traditional data modeling approaches by focusing on simplicity and efficiency of data access. This is important in BI and Big Data environments, where speed and accuracy in gaining insights are essential.

Below we explore how to structure data in an intuitive and effective way, using dimensions and facts to speed up analysis and facilitate understanding.

```python
# Importing all the necessary libraries:
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
pd.set_option('display.max_columns', None)
```

```python
url = 'https://raw.githubusercontent.com/sushantag9/Supermarket-Sales-Data-Analysis/refs/heads/master/supermarket_sales%20-%20Sheet1.csv'

# Importing the DataFrame and displaying it:
df = pd.read_csv(url)
df
```

```python
# Mapping the kinds of products into a numerical value and generating a new column for colors:
unique_product_types = df['Product line'].unique()
product_type_colors = range(len(unique_product_types))
colors_map = dict(zip(unique_product_types, product_type_colors))
df['Product line color'] = df['Product line'].map(colors_map)

# Creating the chart:
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot:
scatter = ax.scatter(
    df['Unit price'],
    df['Quantity'],
    df['Total'],
    c=df['Product line color'],  # Color based in the product type
    s=50,                       # Fixed size of points, can be ajusted accordingly
    cmap='viridis',             # Map of colors
    alpha=0.6,
    edgecolors='w'
)

# Titles and labels:
ax.set_xlabel('Unit Price')
ax.set_ylabel('Quantity')
ax.set_zlabel('Total')

# Captions for the types of products:
legend1 = ax.legend(*scatter.legend_elements(), title="Linha de Produto")
ax.add_artist(legend1)

# Creating captions based in the product types names:
handles, labels = scatter.legend_elements(prop='colors', alpha=0.6)
legend2 = ax.legend(handles, unique_product_types, title='Linha de Produto', loc='upper left')

# Displaying the chart:
plt.show()
```
